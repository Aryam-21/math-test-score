# Mathmatics score prediction

```python
import pandas as pd
from pandas import DataFrame
import  matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('9.1 lsd_math_score_data.csv')
data.head()
```

```python
time = DataFrame(data, columns=['Time_Delay_in_Minutes'])
lsd = DataFrame(data, columns=['LSD_ppm'])
score = DataFrame(data, columns=['Avg_Math_Test_Score'])
```

```python
model = LinearRegression()
model.fit(lsd, score)
pred_score = model.predict(lsd)
```

```python
plt.scatter(lsd, score, color='blue', label='Actual Scores')
plt.plot(lsd, pred_score, color='red', label='Predicted Scores')
plt.title('LSD vs Math Test Scores')
plt.xlabel('LSD ppm')
plt.ylabel('Average Math Test Score')
plt.show()
```
